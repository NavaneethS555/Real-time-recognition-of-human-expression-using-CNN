```python
# OpenCV provides a real-time optimized Computer Vision library, tools, and hardware.
# cv2 is a module within OpenCV
import cv2
```

```python
img1 = cv2.imread("h3.jpg")
img2 = cv2.imread("sad_img1.jpg")
```

```python
# Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python
# It is used to plot the graph
import matplotlib.pyplot as plt
```

```python
plt.imshow(img1)
```

```python
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(img2)
```

```python
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
```

```python
# Deepface is a lightweight face recognition and facial attribute analysis (age, gender, emotion and race) framework for python.
from deepface import DeepFace
```

```python
prediction1 = DeepFace.analyze(img1, enforce_detection=False)
```
